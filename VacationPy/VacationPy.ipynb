```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

```python
# Store Part I results into DataFrame
# Load the csv exported in Part I to a DataFrame
weather_data = pd.read_csv('../WeatherPy/weatherdata.csv', low_memory=False)
weather_data
```

```python
clean_weather = weather_data.dropna()
clean_weather
```

```python
# Humidity Heatmap
# Configure gmaps.
gmaps.configure(api_key=g_key)
# Use the Lat and Lng as locations and Humidity as the weight.
locations = clean_weather[['Latitude', 'Longitude']]
humidity = clean_weather['Humidity (%)'].astype(float)
# Add Heatmap layer to map.
center_coord = (0,0)
figure_layout = {
    'width': '800px',
    'height': '500px',
    'border': '1px solid black',
    'padding': '1px'
}
fig = gmaps.figure(center=center_coord, zoom_level=1.5, layout=figure_layout)
heat_layer = gmaps.heatmap_layer(locations, weights=humidity)
fig.add_layer(heat_layer)
fig

```

```python
# Create new DataFrame fitting weather criteria
# Narrow down the cities to fit weather conditions.
ideal_temp = clean_weather.loc[(clean_weather['Temperature (F)']>75) & (clean_weather['Temperature (F)']<80),:]
ideal_wind=ideal_temp.loc[(ideal_temp['Windspeed (mph)']<5)]
ideal_weather = ideal_wind.loc[(ideal_wind['Cloudiness (%)'] <10)]
ideal_weather
```

```python
# Hotel Map
# Store into variable named hotel_df.
hotel_df = ideal_weather
# Add a "Hotel Name" column to the DataFrame.
hotel_df['Hotel Name'] = ""
hotel_df['Country']=''
hotel_df
```

```python
# Set parameters to search for hotels with 5000 meters.
params = {"radius" : 5000,
          "types" : "hotel",
          "key": g_key    
}
for index, row in hotel_df.iterrows():
    lat=row['Latitude']
    lng=row['Longitude']
    params["location"] = f"{lat}, {lng}"
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    name_address = requests.get(base_url, params=params)
    name_address=name_address.json()
    #print(json.dumps(name_address, indent = 4, sort_keys=True))
# Hit the Google Places API for each city's coordinates.
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address['results'][1]['name']
        hotel_df.loc[index, "Country"] = name_address['results'][1]['plus_code']['compound_code'].split(', ')[-1]
    except:
        print(row['City'] + "does not have a nearby hotel")
# Store the first Hotel result into the DataFrame.        
hotel_df
```

```python
hotel_df=hotel_df.rename(columns={"Latitude": "Lat", "Longitude":"Lng"})
#plot markers on top of the heatmap
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
fig
```
